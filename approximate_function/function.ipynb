### Описываем функцию

```python
import math
def f (x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
```

### Импортируем необходимые библиотеки

```python
import numpy
from matplotlib import pylab
from scipy import linalg 
```

### Составим СЛУ для многочлена первой степени, решим и посмотроим график

```python
A = numpy.array([[1, 1],[1, 15]])
b = numpy.array([[f(1)],[f(15)]])
solve = linalg.solve(A,b)
print(solve)
print([(solve[0]*x+solve[1]*x) for x in numpy.arange(0, 15, 0.1)])
pylab.plot(numpy.arange(0, 15, 0.1), [f(x) for x in numpy.arange(0, 15, 0.1)])
pylab.plot(numpy.arange(0, 15, 0.1), [(solve[0]+solve[1]*x) for x in numpy.arange(0, 15, 0.1)])
```

### Составим СЛУ для многочлена второй степени, решим и посмотроим график

```python
A = numpy.array([[1,1,1],[1, 8, 64],[1, 15, 255]])
b = numpy.array([[f(1)],[f(8)],[f(15)]])
solve = linalg.solve(A,b)
print(solve)
pylab.plot(numpy.arange(0, 15, 0.1), [f(x) for x in numpy.arange(0, 15, 0.1)])
pylab.plot(numpy.arange(0, 15, 0.1), [(solve[0]+solve[1]*x+solve[2]*x*x) for x in numpy.arange(0, 15, 0.1)])
```

### Составим СЛУ для многочлена третьей степени, решим и посмотроим график

```python
A = numpy.array([[1, 1, 1, 1], [1, 4, 4**2, 4**3], [1, 10, 10**2, 10**3], [1, 15, 15**2, 15**3]])
b = numpy.array([[f(1)],[f(4)],[f(10)],[f(15)]])
solve = linalg.solve(A,b)
print(solve)
pylab.plot(numpy.arange(0, 15, 0.1), [f(x) for x in numpy.arange(0, 15, 0.1)])
pylab.plot(numpy.arange(0, 15, 0.1), [(solve[0]+solve[1]*x+solve[2]*x*x+solve[3]*x*x*x) for x in numpy.arange(0, 15, 0.1)])
```

```python
with open('function_result.txt', 'w') as file_out:
    answer = " ".join(map(lambda x: str(round(x[0],2)), solve))
    file_out.write(answer)
```
